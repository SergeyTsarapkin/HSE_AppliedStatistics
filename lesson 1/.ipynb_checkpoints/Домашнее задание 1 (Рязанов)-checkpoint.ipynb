### 1. Рассмотрим выборку x1, . . . , xn из экспоненциального распределения с параметром θ. Ответьте на следующие вопросы:


#### • (5 баллов) Найдите оценку методом моментов для параметра θ (теоретически, как показано на слайдах лекции).

#### Ответ:

Знаем, что плотность экспоненциального распределения выглядит следующим образом

$f(x;λ) = λe^{-λx}, x>=0 $

Поскольку у нашего распределения один параметр λ, оценка параметра θ фактически приводит к оценке параметра λ. $f(x;θ) = θe^{-θx} $

Тогда, сначала найдем первый выборочный момент:

$\frac{1}n\sum \limits _{j=1}^{n}x_{j} = \int_{-\infty}^\infty xf(x;θ)\:dx = \
\int_0^\infty xθe^{-λx}\:dx = \frac{1}θ$

Отсюда получаем оценку параметра θ:

$\hat{θ} = \frac{1}{\frac{1}n\sum \limits _{j=1}^{n}x_{j}} = \frac{n}{\sum \limits _{j=1}^{n}x_{j}} = \
\frac{1}{\bar x}$

#### • (5 баллов) Найдите оценку максимального правдоподобия для параметра θ (теоретически, как показано на слайдах лекции).

#### Ответ:

Функция правдоподобия выглядит следующим образом:

$L(θ,x_1,...,x_n) = \prod \limits _{j=1}^{n}f(x_{j},θ) = \
\prod \limits _{j=1}^{n}θe^{-θx} = θ^{n}exp(-θ\sum \limits _{j=1}^{n}x_{j})$

Чтобы найти оценку максимального правдоподобия нам нужно прологарифмировать функцию правдоподобия, взять от нее производную и приравнять к 0:

$\frac{\:d ln(L(θ,x_1,...,x_n))}{\:d θ} = 0$

$\frac{\:d ln(L(θ,x_1,...,x_n))}{\:d θ} = \frac{\:d \
ln(θ^{n}exp(-θ\sum \limits _{j=1}^{n}x_{j})}{\:d θ} = \
\frac{\:d \
ln(nln(θ) - exp(-θ\sum \limits _{j=1}^{n}x_{j})}{\:d θ} = \
\frac{n}{θ} - \sum \limits _{j=1}^{n}x_{j}$

Отсюда оценка параметра $\hat{θ}$:

$\hat{θ} = \frac{n}{\sum \limits _{j=1}^{n}x_{j}} = \
\frac{1}{\bar x}$

#### • (15 баллов) Визуально (как на практическом занятии) проверьте полученные оценки на несмещённость и состоятельность (зафиксировав произвольное значение параметра θ и построив график зависимости значения оценки от размера выборки).

#### Ответ:

Для визуальной оценки несмещенности (при фиксированном размере выборки мы в среднем не ошибаемся) и состоятельности (при бесконечном расширении выборки мы приходим к истинному значению) сделаем следующее:
1. Покажем сравнение истинного значения параметра 1/θ, среднего значения генеральной совокупности (заданного размера) и среднего значения выборки.
2. Построим график оценки параметра в зависимости от количества наблюдений.

```python
import numpy as np
import pandas as pd

import scipy.stats as sps
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.rcParams['figure.dpi'] = 80
%matplotlib inline

my_theta = 7

expon_rv = sps.expon(scale=1/my_theta)  
print(f'Истинное значение параметра 1/(θ={my_theta}) = {round(1/my_theta,6)}')

# Представим, что вся генеральная совокупность имеет размер 10^5 
size = 10**5
x_test = expon_rv.rvs(size = size)

real_mean = np.mean(x_test) # среднее генеральной совокупности
print(f'Среднее значение генеральной совокупности (размера {size}) \
= {round(real_mean,6)}')

n = 300
number_of_samples = 400

th_od = np.zeros(number_of_samples)

for i in range(number_of_samples):
    x_sample = np.random.choice(x_test, size=n, replace=False)
    x_mean = np.mean(x_sample)
    th_od[i] = x_mean

print(f'Среднее значение выборки = {round(np.mean(th_od),6)}')

theta_hat_od = []
for n in range(100, 10000, 100): 
    
    # 100 выборок с повторениями
    x_sample = np.random.choice(x_test, size = (n, 100))
    
    # 100 разных оценок 
    x_mean = np.mean(x_sample, axis=0) 
    
    theta_hat_od.append(x_mean)               

plt.figure(figsize=(12,8))
plt.title("График оценки", fontsize=15)
plt.plot(range(100, 10000, 100), theta_hat_od, alpha = 0.1)

plt.xlabel('Количество наблюдений', size=15)
plt.ylabel('Оценка параметра', size=15)

plt.hlines(real_mean, 100, 10000, color='red', lw=1, \
           label=f'среднее генеральной совокупности = {round(real_mean,6)}')
plt.hlines(1/my_theta, 100, 10000, color='blue', lw=1, \
           label=f'истинное значение параметра 1/θ = {round(1/my_theta,6)}')
plt.hlines(np.mean(th_od), 100, 10000, color='green', lw=1, \
           label=f'среднее выборки = {round(np.mean(th_od),6)}')
plt.legend(fontsize=15);    
```

В результате визуальной проверки можно сказать, что полученные оценки параметра несмещенные и состоятельные.

#### • (2 балла) Для какого параметра статистика $(\bar x)^{2}$ будет являться оценкой максимального правдоподобия? Каким свойством ML-оценки вы воспользовались для ответа на данный вопрос?

#### Ответ:

Воспользуемся свойством ML-оценки, который называется инвариантностью:

Зная, что $\hat{θ} = \frac{1}{\bar x}$ и $θ^{MLE} = (\bar x)^{2}$ получим:

$\frac{1}{\bar x} = (\bar x)^{2}$ =>
$1 = (\bar x)^{3}$ =>
$\bar x = 1$, отсюда => $\hat{θ} = \frac{1}{\bar x} = \
\frac{1}{1} = 1$

#### • (15 баллов) Зафиксируйте произвольное значение параметра θ и для выборок размера 10, 100, 1000 промоделируйте распределение оценок $(\bar x)^{2}$ и $\frac{\sum \limits _{i=1}^{n}x_{i}^{2}}{2n}$. Постройте гистограммы.

```python
my_theta = 3

expon_rv = sps.expon(scale=1/my_theta)  
```

```python
obs_10 = 10
obs_100 = 100
obs_1000 = 1000

x = expon_rv.rvs(size = (obs_10, 1000))
y = expon_rv.rvs(size = (obs_100, 1000))
z = expon_rv.rvs(size = (obs_1000, 1000))
print(x.shape)
print(y.shape)
print(z.shape)
```

```python
# Выборка 10 
theta_x_1 = np.mean(x, axis=0)*np.mean(x, axis=0)
theta_x_2 = np.sum(x*x, axis=0)/(2*obs_10)
# Выборка 100  
theta_y_1 = np.mean(y, axis=0)*np.mean(y, axis=0)
theta_y_2 = np.sum(y*y, axis=0)/(2*obs_100)
# Выборка 1000  
theta_z_1 = np.mean(z, axis=0)*np.mean(z, axis=0)
theta_z_2 = np.sum(z*z, axis=0)/(2*obs_1000)
```

```python
print(theta_x_1.mean(), theta_y_1.mean(), theta_z_1.mean())
print(theta_x_2.mean(), theta_y_2.mean(), theta_z_2.mean())
```

#### • (15 баллов) Оцените методом моментов или максимального правдоподобия параметры распределения. Проверьте, что посчитанные параметры соответствуют действительности: постройте график плотности с полученными параметрами, наложите его на гистограмму выборки.

#### Ответ:
Оценив методом максимального правдоподобия параметры распределения мы получим, что параметр среднего значения определяется по формуле:
$\hat μ_n = \frac{1}{n}\sum \limits _{j=1}^{n}x_{j}$, а параметр дисперсии определяется по формуле:
$\hat σ_n = \frac{1}{n}\sum \limits _{j=1}^{n}(x_{j}-\hat μ)^{2}$

Теперь проверим, что посчитанные параметры соответствуют
действительности:


```python
obs = 1000
theta_mean = (1/obs)*np.sum(data, axis=0)

theta_var = (1/obs)*np.sum((data-theta_mean)*(data-theta_mean), axis=0)
```

```python
# Оценочные параметры
theta_mean, theta_var 
```

```python
# Рассчитанные из выборки параметры
my_mean, my_var
```

Очевидно, что оценочные параметры совпали с параметрами, рассчитанными из выборки. Построим график плотности с полученными параметрами и наложим его на гистограмму выборки:

```python
norm_rv = sps.norm(loc=theta_mean, scale=np.sqrt(theta_var))
x = np.linspace(0, 22, 100)
pdf = norm_rv.pdf(x)

# плотность
plt.plot(x, pdf, lw=3, color='m')

# гистограмма, параметр density отнормировал её. 
sns.histplot(data, bins=20,stat='density')

plt.ylabel('$f(x)$')
plt.xlabel('$x$');
```

Таким образом, можно уверенно сказать, что данные из выборки принадлежат нормальному распределению.

### 3 (10 баллов) На первой лекции мы увидели, что в зависимости от распределения, матожидание может как совпадать с медианой, так и находится левее или правее от неё. Но насколько далеко может отстоять матожидание от медианы? Для этого ответьте на вопрос: может ли матожидание находится левее первого квартиля? Если да, приведите пример распределения (дискретного или непрерывного), для которого это верно. Если нет, объясните, почему.

#### Ответ:
Матожидание, безусловно, может находиться левее первого квартиля или правее третьего квартиля (то есть находиться снаружи интерквартильного размаха). Причиной тому служит тот факт, что квартили устойчивы к изменениям в данных и выбросам, в то время, как значение математического ожидания очень чувствительно к такого рода ситуациям. 

Когда мы добавляем ϵ к любому отдельному значению в наборе данных из n > 4 чисел, матожидание изменяется на величину ϵ/n, которое может быть сколь угодно большим. Однако, квартили (если вообще меняются) сдвигаются только к соседним значениям в исходном наборе данных и, таким образом, ограничены в том, насколько они могут измениться.

Рассмотрим следующий пример:

```python
my_dataset = (1,2,2000000,2000001,2000002,2000003,2000004,2000005,2000006,2000007)
print(f'Возьмем набор данных: {my_dataset}')
print(f'Первый квартиль = {np.quantile(my_dataset, .25)}')
print(f'Третий квартиль = {np.quantile(my_dataset, .75)}')
print(f'Среднее значение = {np.mean(my_dataset)}')
```

Таким образом, распределение с большими выбросами может легко спровоцировать ситуацию, когда матожидание будет находиться за пределами интерквартильного диапазона, или в нашем случае, находиться левее первого квартиля.

Теперь попробуем отетить на вопрос насколько далеко матожидание может отстоять от медианы:

Зная, что $E(|X−c|)$ минимальна, когда $c = M(X)$, а также что $E[g(X)] \geq g(E[X])$ (неравенство Йенсена), получим:

$|M(X) - E(X)| = |E(M(X) - X| \leq E(|X - M(X)|) \leq E(|X - E(X)|) \
\leq \sqrt{E(|X - E(X)|^{2})} \leq (Var(X))^{1/2}$, а именно
$|M(X) - E(X)| \leq σ$ 

Получается, что дистанция между матожиданием и медианой не может превышать стандартного отклонения. Убедимся, что это неравенство выполняется и в нашем случае:

```python
print(f'Медиана - Матожидание = {np.quantile(my_dataset, .50)-np.mean(my_dataset)}')
print(f'Стандартное отклонение = {np.sqrt(np.var(my_dataset))}')
```
