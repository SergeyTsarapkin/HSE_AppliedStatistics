### 1) Выборка $ x_{1},....,x_{n} $ из экспоненциального распределения с параметром $ \theta $

#### Оценка параметра $\theta$ методом моментов (приравняем первые моменты выборки и самого распределения):</br>

Первый момент экспоненциального распределения (из определения): $\mathbb{E}(x) = \frac{1}{\theta}$ </br>
Первый момент выборки: $\mathbb{E}(x) = \frac{\sum \limits _{i=1} ^{n} x_{i}}{n} = \bar{x}$ </br></br>
Получаем: $\frac{1}{\theta} = \bar{x}\ \Rightarrow\ \theta=\frac{1}{\bar{x}}$, где $\bar{x} = $ выборочное среднее </br>
##### Ответ: методом моментов  $\hat{\theta}_{MM} = \frac{1}{\bar{x}}$
***

#### Оценка параметра $\theta$ методом максимального правдоподобия (найти максимум для $\theta$):</br>

$L(\theta | x_{1},...x_{n})=\prod_{i=1}^{n}f(x_{i}|\theta)= \theta \mathbb{e}^{-\theta x_{1}} \cdot \theta \mathbb{e}^{-\theta x_{2}}\cdot\cdot\cdot \theta \mathbb{e}^{-\theta x_{n}} = \theta^{n}\mathbb{e}(-\theta \sum \limits _{i=1}^{n}x_{i}) \\ $

Для удобства поиска максимума прологарифмируем: 

$ \ln(\theta^{n}\mathbb{e}(-\theta \sum \limits _{i=1}^{n}x_{i})) = n\ln(\theta) - \theta \sum \limits_{i=1}^{n}x_{i}$

Найдем максимум через производную, приравняв ее к нулю: </br>
$\frac{\partial}{\partial \theta}(n\ln(\theta) - \theta \sum \limits_{i=1}^{n}x_{i}) = \frac{n}{\theta} - \sum \limits_{i=1}^{n}x_{i} \\ $

$\frac{n}{\theta} - \sum \limits_{i=1}^{n}x_{i} = 0\ \Rightarrow\ \frac{n}{\theta} = \sum \limits_{i=1}^{n}x_{i}\ \Rightarrow\ \theta = \frac{1}{\frac{1}{n}\sum \limits_{i=1}^{n}x_{i}}\ \Rightarrow\ \theta = \frac{1}{\bar{x}}$ где $\bar{x} = $ выборочное среднее

##### Ответ: методом максимального правдоподобия $\hat{\theta}_{ML} = \frac{1}{\bar{x}}$
***

#### Несмещенность и состоятельность

```python
fixed_theta = 2.5 #зафиксировали произвольное значение
```

```python
rasp = expon(scale=1/fixed_theta)
gensov = rasp.rvs(size=1_000_000) #Генеральная совокупность
real_mean = gensov.mean() #Настоящее среднее 
```

```python
number_of_samples = 300
sample_size = 200
sample_means = np.zeros(number_of_samples)

for i in range(number_of_samples):
    sample = np.random.choice(gensov, size=sample_size, replace=False)
    mean = np.mean(sample)
    sample_means[i] = mean
    
print(f'Настоящее среднее генеральной совокупности = {real_mean}')
print(f'Среднее значение по {number_of_samples} выборкам = {sample_means.mean()}')
```

```python
theta_hat = []
for i in range(100, 10000, 100):
    sample = np.random.choice(gensov, size = (i, 100))
    mean = np.mean(sample, axis=0)
    theta_hat.append(mean)
```

```python
plt.figure(figsize=(14,10))
plt.plot(range(100, 10000, 100), theta_hat, alpha = 0.1);

plt.xlabel('Кол-во наблюдений', size = 17);
plt.ylabel('Оценка праметра', size = 17);

plt.hlines(real_mean, 100, 10000, color = 'red', lw=2, \
           label=f'настоящее среднее (из ген. совокупности) = {round(real_mean,4)}')
plt.hlines(1/fixed_theta, 100, 10000, color='blue', lw=2, \
           label=f'Истинная ( зафиксированная) theta = {round(1/fixed_theta,4)}')
plt.hlines(np.mean(sample_means), 100, 10000, color='green', lw=2, \
           label=f'среднее выборки = {round(np.mean(sample_means),6)}')
plt.legend(fontsize=15);
plt.title('График оценки', fontsize=17);
```

Ответ: полученные оценки параметра являются несмещенными и состоятельными.

#### Для какого параметра статистика $(\bar x)^{2}$ будет являться оценкой максимального правдоподобия? Каким свойством ML-оценки вы воспользовались для ответа на данный вопрос?

Инвариативность: </br> 
$\hat{\theta} = \frac{1}{\bar x}$ , $\theta^{MLE} = (\bar x)^{2}$

$\frac{1}{\bar x} = (\bar x)^{2} \Rightarrow 1 = (\bar x)^{3} \Rightarrow \bar x = 1$ </br>
Соответственно: </br>
$\hat{\theta} = \frac{1}{\bar x} = \frac{1}{1} = 1$
***

#### Зафиксируйте произвольное значение параметра θ и для выборок размера 10, 100, 1000 промоделируйте распределение оценок $(\bar x)^{2}$ и $\frac{\sum \limits _{i=1}^{n}x_{i}^{2}}{2n}$. Постройте гистограммы.

```python
theta = 5.5
rasp = expon(scale=1/theta)

observs_10, observs_100, observs_1000 = 10, 100, 1000
x = rasp.rvs(size=(observs_10, 1000))
y = rasp.rvs(size=(observs_100, 1000))
z = rasp.rvs(size=(observs_1000, 1000))
```

```python
theta_x_a = np.mean(x, axis=0)*np.mean(x, axis=0)
theta_x_b = np.sum(x*x, axis=0)/(2*observs_10)

theta_y_a = np.mean(y, axis=0)*np.mean(y, axis=0)
theta_y_b = np.sum(y*y, axis=0)/(2*observs_100)

theta_z_a = np.mean(z, axis=0)*np.mean(z, axis=0)
theta_z_b = np.sum(z*z, axis=0)/(2*observs_1000)
```

##### 1) Гистограмма для $(\bar x)^{2}$

```python
plt.figure(figsize=(14,10))
plt.hist(theta_x_a, bins=100, alpha=0.5, label='10 выборок', color='r')
plt.hist(theta_y_a, bins=50, alpha=0.5, label='100 выборок',color='g')
plt.hist(theta_z_a, bins=20, alpha=0.5, label='1000 выборок', color='b')
plt.xlabel('Оценка параметра', size=17)
plt.ylabel('Количество наблюдений', size=17)
plt.legend(fontsize=17);
```

##### 2) Гистограмма для $\frac{\sum \limits _{i=1}^{n}x_{i}^{2}}{2n}$

```python
plt.figure(figsize=(14,10))
plt.hist(theta_x_b, bins=100, alpha=0.5, label='10 выборок', color='r')
plt.hist(theta_y_b, bins=50, alpha=0.5, label='100 выборок', color='g')
plt.hist(theta_z_b, bins=20, alpha=0.5, label='1000 выборок', color='b')
plt.xlabel('Оценка параметра', size=17)
plt.ylabel('Количество наблюдений', size=17)
plt.legend(fontsize=17);
```

#### Оцените методом моментов или максимального правдоподобия параметры распределения. Проверьте, что посчитанные параметры соответствуют действительности: постройте график плотности с полученными параметрами, наложите его на гистограмму выборки.

##### Методом максимального правдоподобия мы получим: </br>
$\hat μ_n = \frac{1}{n}\sum \limits _{j=1}^{n}x_{j} \\ \hat σ_n = \frac{1}{n}\sum \limits _{j=1}^{n}(x_{j}-\hat μ)^{2}$



```python
observs = 1000
theta_mean = (1/observs)*np.sum(data, axis=0)
theta_var = (1/observs)*np.sum((data-theta_mean)*(data-theta_mean), axis=0)
theta_mean, theta_var 
```

```python
#Расчитынные ранее параметры по выборке:
mean, var
```

```python
#Разница
mean-theta_mean, var-theta_var
```

```python
plt.figure(figsize=(14,10))
norm_rv = norm(loc=theta_mean, scale=np.sqrt(theta_var))
x = np.linspace(0, 21, 100)
pdf = norm_rv.pdf(x)

plt.plot(x, pdf, lw=2, color='r')

sns.histplot(data, bins=40,stat='density')
```

##### Ответ: данные из выборки с нормальным распределением
